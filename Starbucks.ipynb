```python
df_starbucks['Vitamin A (% DV) '] = df_starbucks['Vitamin A (% DV) '].str.replace('%', ' ').astype(int)
df_starbucks['Vitamin C (% DV)'] = df_starbucks['Vitamin C (% DV)'].str.replace('%', ' ').astype(int)
df_starbucks[' Calcium (% DV) '] = df_starbucks[' Calcium (% DV) '].str.replace('%', ' ').astype(int)
```

```python
df_starbucks['Iron (% DV) '] = df_starbucks['Iron (% DV) '].str.replace('%', ' ').astype(float)
```

```python
df_starbucks['Caffeine (mg)'] = df_starbucks['Caffeine (mg)'].str.replace('%', ' ')
df_starbucks['Caffeine (mg)'] = df_starbucks['Caffeine (mg)'].replace('Varies', 
    np.mean(pd.to_numeric(df_starbucks['Caffeine (mg)'], errors='coerce')))
df_starbucks['Caffeine (mg)'] = pd.to_numeric(df_starbucks['Caffeine (mg)'], errors='coerce').pipe(lambda s: s.fillna(s.mean()))
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
plt.title('Vitamin A and C')
plt.xticks(rotation=90)
 
# using the twinx() for creating another
# axes object for secondary y-Axis
ax2 = ax.twinx()
ax.scatter(data=df_starbucks,x="Beverage_category", y='Vitamin A (% DV) ', color = 'g')
ax2.scatter(data=df_starbucks,x="Beverage_category",y='Vitamin C (% DV)', color = 'b')
 
# giving labels to the axises
ax.set_xlabel('Beverage_Category')
ax.set_ylabel('Vitamin A (% DV) ')
ax.legend()
# secondary y-axis label
ax2.set_ylabel('Vitamin C (% DV)', color = 'b')
 
# defining display layout
plt.tight_layout()
 
# show plot
plt.show()
```
